```python
#Exploratory Data Analysis (EDA) and Visualization:
# Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='Acquisition_Cost' ,y='ROI',data=df)                                #Scatter Plot (à®ªà¯à®³à¯à®³à®¿ à®µà®°à¯ˆà®ªà®Ÿà®®à¯) à®‰à®°à¯à®µà®¾à®•à¯à®•à¯à®®à¯ à®šà¯†à®¯à®²à¯à®ªà®¾à®Ÿà¯.
plt.title('Acquisition_Cost vs ROI',fontsize='10',fontweight='10',color='red')
plt.show()

```

```python
#Create a bar chart to visualize the average Conversion_Rate for different
#Channel_Used, categorized by Campaign_Type:

plt.figure(figsize=(10,6))
sns.barplot(x='Channel_Used',y='Conversion_Rate',data=df,hue='Campaign_Type',estimator='mean' )  #(subgroups: Email Campaign, Social Media Campaig)
plt.title('Channel_Used')
plt.ylim(0,10)                                              # Y-axis(limit(0,10))
plt.show()
```

```python
#Visualize the distribution of Engagement_Score across different Campaign_Type
#using a box plot.
plt.figure(figsize=(10,8))
plt.xticks(rotation=45)
plt.title('Visualize the distribution of Engagement_Score',fontsize='10',fontweight='bold',color='green')
sns.boxplot(x='Campaign_Type',y='Engagement_Score',data=df,hue='Campaign_Type')
plt.show()

```

```python
#Analyze the average ROI by Company using a bar chart to compare the profitability
#of campaigns conducted by different companies.

plt.figure(figsize=(12,5))
plt.title('Analyze the average ROI by Company using a bar chart')
sns.barplot(x='Company',y='ROI',estimator ='mean',data=df, hue='Campaign_Type')
plt.show()
```

```python
#Examine the correlation between Engagement_Score and Conversion_Rate
#using a heatmap.
correlation=df[['Engagement_Score','Conversion_Rate']].corr()   # correlation coefficient (.corr)
print(correlation)
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

```python
#Customer Segmentation:
#Create a count plot to visualize the distribution of Target_Audience.
plt.figure(figsize=(10,8))
plt.title('the distribution of Target_Audience',fontsize='10',fontweight='15',color='blue')
x=sns.countplot(x='Target_Audience',data=df,color='orange')
x.bar_label(x.containers[0])
plt.show()
```

```python
#Identify which Customer_Segment has the highest Conversion_Rate for each
#Language using a bar chart.
```

```python
plt.figure(figsize=(16,6))
plt.ylim(0,9)
plt.title('Customer_Segment has the highest Conversion_Rate',fontweight='bold' ,color='green')
x=sns.barplot(x='Customer_Segment',y='Conversion_Rate',data=df,estimator='max',hue_order='language',width=0.9)
for container in x.containers:
   x.bar_label(container,fmt='%.2f')
plt.show()
```

```python
#Visualize the distribution of Acquisition_Cost across each Customer_Segment,
#categorized by Channel_Used, using a box plot.
plt.figure(figsize=(15,5))
plt.xticks(rotation=45)
plt.title(' the distribution of Acquisition_Cost across each Customer_Segmen',fontsize='10',fontweight='bold',color='green')
sns.boxplot(x='Customer_Segment',y='Acquisition_Cost',data=df,hue='Channel_Used')
plt.show()
```

```python
#Analyze average Conversion_Rate by Language using a bar chart to compare
#the effectiveness of campaigns conducted in different languages.
```

```python
plt.figure(figsize=(13,7))
plt.xticks(rotation=45)
plt.title('Analyze average Conversion_Rate by Language using a bar chart',fontsize='10',fontweight='bold',color='orange')
ax=sns.barplot(x='Language',y='Conversion_Rate',data =df,hue='Campaign_Type',width=0.9)
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f', padding=4)
plt.show()
```

```python
#Channel Effectiveness:
#Compare the Engagement_Score for different Channels_Used, segmented by
#Campaign_Type, using a bar chart.

plt.figure(figsize=(15,5))
plt.xticks(rotation=45)
plt.title('the Engagement_Score for different Channels_Used, segment',fontsize='10',fontweight='bold',color='green')
sns.barplot(x='Channel_Used',y='Engagement_Score',data=df,hue='Campaign_Type',palette='Set2')
plt.show()
```

```python
#Show the distribution of total ROI across different Channels_Used using a pie chart:
pie=df.groupby('Channel_Used')['ROI'].sum()
plt.title('the distribution of total ROI across different Channels_Used using a pie chart')
plt.pie(pie,labels=(pie.index),autopct='%1.2f%%',shadow=True)
plt.show()
```

```python
#Plot a scatter plot to show the relationship between Clicks and Impressions
#for each Channel_Used.
plt.figure(figsize=(8,7))
plt.title(' the relationship between Clicks and Impressions by channel used',fontweight='bold',color='red')
sns.scatterplot(x='Clicks',y='Impressions',data=df,hue='Channel_Used')
plt.show()
```

```python
#Time-Based Analysis:
#Plot the distribution of Duration using a histogram.
plt.figure(figsize=(10,5))
plt.title('distribution of Duration')
sns.histplot(x='Duration',data=df,color='green',kde=True)  #KDE (Kernel Density Estimate)
plt.show()
```

```python
#Analyze how the overall Conversion_Rate has changed over Date for each
#Company using a line chart.
plt.figure(figsize=(10,5))
plt.title('Analyze how the overall Conversion_Rate has changed over Date',fontweight='bold',color='red')
sns.lineplot(x='Date',y='Conversion_Rate',data=df,hue='Company',errorbar=None) #KDE (Kernel Density Estimate)
plt.show()

```

```python
#Examine the trend of Engagement_Score over Date with a line chart.
plt.figure(figsize=(10,5))
plt.title('the trend of Engagement_Score over Date with a line chart',fontweight='bold',color='green')
sns.lineplot(x='Date',y='Engagement_Score',data=df,hue='Company',errorbar=None,linestyle='--',marker='.') #KDE (Kernel Density Estimate)
plt.show()

```

```python
#Geographic Analysis:
#Determine which location has the highest Acquisition_Cost using a bar chart.

plt.figure(figsize=(13,7))
plt.xticks(rotation=45)
plt.title('location has the highest Acquisition_Cost',fontsize='10',fontweight='bold',color='green')
ax=sns.barplot(x='Location',y='Acquisition_Cost',data =df,hue='Location')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
#Visualize the Conversion_Rate by different Location, categorized by
#Target_Audience, using a bar chart.
plt.figure(figsize=(13,7))
plt.xticks(rotation=45)
plt.title('Visualize the Conversion_Rate by different Location, categorized',fontsize='10',fontweight='bold',color='red')
ax=sns.barplot(x='Location',y='Conversion_Rate',data =df,hue='Target_Audience',palette='Set3')

plt.show()
```

```python
#Illustrate the proportion of ROI by Location using a pie chart.
proportion=df.groupby('Location')['ROI'].sum()
#proportion
plt.figure(figsize=(10,5))
plt.title('proportion of ROI by Location using a pie chart',fontweight='bold',color='green')
plt.pie(proportion,labels =proportion.index ,autopct='%1.2f%%',shadow=True)
plt.show()
```
